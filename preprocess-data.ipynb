```python
import pandas as pd
```

```python
df_b = pd.read_json("dataset/yelp_academic_dataset_business.json", lines=True)
```

```python
df_b.shape
```

```python
df_b = df_b[df_b['is_open']==1]
drop_columns = ['hours','is_open','review_count']
df_b = df_b.drop(drop_columns, axis=1)
```

```python
df_b.head()
```

```python
df_explode = df_b.assign(categories = df_b.categories
                         .str.split(', ')).explode('categories')
df_explode.categories.value_counts()
```

```python
df_restaurants = df_b[df_b['categories'].str.contains('Restaurants|Food', case=False, na=False)]
```

```python
df_restaurants
```

```python
size = 1000000
review = pd.read_json("yelp_academic_dataset_review.json", lines=True,
                      dtype={'review_id':str,'user_id':str,
                             'business_id':str,'stars':int,
                             'date':str,'text':str,'useful':int,
                             'funny':int,'cool':int},
                      chunksize=size)
```

```python
chunk_list = []
for chunk_review in review:
    # Drop columns that aren't needed
    chunk_review = chunk_review.drop(['review_id','useful','funny','cool'], axis=1)
    # Renaming column name to avoid conflict with business overall star rating
    chunk_review = chunk_review.rename(columns={'stars': 'review_stars'})
    # Inner merge with edited business file so only reviews related to the business remain
    chunk_merged = pd.merge(df_restaurants, chunk_review, on='business_id', how='inner')
    # Show feedback on progress
    print(f"{chunk_merged.shape[0]} out of {size:,} related reviews")
    chunk_list.append(chunk_merged)
# After trimming down the review file, concatenate all relevant data back to one dataframe
df = pd.concat(chunk_list, ignore_index=True, join='outer', axis=0)
```

```python
csv_name = "yelp_reviews_restaurants.csv"
df.to_csv(csv_name, index=False)
```

```python
df_subset = df.iloc[:250000,:]
```

```python
csv_name = "yelp_reviews_restaurants_subset.csv"
df_subset.to_csv(csv_name, index=False)
```

```python
size = 1000000
review = pd.read_csv("yelp_reviews_restaurants.csv", chunksize=size)
```

```python
chunk_list = []
for chunk_review in review:
    chunk_list.append(chunk_review)
# After trimming down the review file, concatenate all relevant data back to one dataframe
df = pd.concat(chunk_list, ignore_index=True, join='outer', axis=0)
```

```python
df.head()
```

```python
def stars2sentiment(x):
  if x == 1 or x == 2: return 0
  elif x == 3: return 1
  elif x == 4 or x == 5: return 2
```

```python
df["sentiment"] = df["review_stars"].apply(stars2sentiment)
```

```python
df.head()
```

```python
df_sampled = df.groupby('sentiment', group_keys=False).apply(lambda x: x.sample(5000))
```

```python
df_sampled.shape
```

```python
csv_name = "yelp_reviews_restaurants_sampled_15000.csv"
df_sampled.to_csv(csv_name, index=False)
```
